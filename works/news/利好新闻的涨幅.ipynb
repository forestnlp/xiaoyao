```python
import pandas as pd

# 读取目录：D:\workspace\xiaoyao\llm\batch_split_results下的industry_verdict_*.parquet
import os

# 定义目录路径
dir_path = "D:/workspace/xiaoyao/llm/batch_split_results"

# 列出目录下所有文件
files = os.listdir(dir_path)

# 筛选出所有industry_verdict_*.parquet文件
parquet_files = [f for f in files if f.startswith("industry_verdict_") and f.endswith(".parquet")]

# 读取每个文件并合并到一个DataFrame中
df_list = []
for parquet_file in parquet_files:
    file_path = os.path.join(dir_path, parquet_file)
    df = pd.read_parquet(file_path)
    df_list.append(df)

# 合并所有DataFrame
merged_df = pd.concat(df_list, ignore_index=True)

# 显示合并后的DataFrame信息
print(merged_df.info())
```

```python
# 获取merged_df里impact_direction为利好且confidence_score>7的记录
good_news_df = merged_df[(merged_df['impact_direction'] == '利好') & (merged_df['confidence_score'] > 8)]
good_news_df.head(3)
```

```python
# 读取目录：D:\workspace\xiaoyao\llm\batch_split_results下的stock_recommendation_*.parquet

import os

# 定义目录路径
dir_path = "D:/workspace/xiaoyao/llm/batch_split_results"

# 列出目录下所有文件
files = os.listdir(dir_path)

# 筛选出所有industry_verdict_*.parquet文件
parquet_files = [f for f in files if f.startswith("stock_recommendation_") and f.endswith(".parquet")]

# 读取每个文件并合并到一个DataFrame中
df_list = []
for parquet_file in parquet_files:
    file_path = os.path.join(dir_path, parquet_file)
    df = pd.read_parquet(file_path)
    df_list.append(df)

# 合并所有DataFrame
merged_stock_df = pd.concat(df_list, ignore_index=True)

# 显示合并后的DataFrame信息
print(merged_stock_df.info())
```

```python
# 使用merged_df 和 merged_stock_df 进行合并
merged_df = pd.merge(merged_df, merged_stock_df, on=['analysis_id'], how='left')
merged_df = merged_df[merged_df['original_news_date'] <= '2025-01-02']
```

```python
# 统计merged_df里stock_code的出现频次，倒序排序
df_recommendation_stock_code = merged_df['stock_code'].value_counts().sort_values(ascending=False)
# 转换为df，打印df_recommendation_stock_code
df_recommendation_stock_code = df_recommendation_stock_code.reset_index()
df_recommendation_stock_code.columns = ['stock_code', 'count']
df_recommendation_stock_code
```

```python
# 统计merged_df里stock_code的出现频次，倒序排序
df_recommendation_industry = merged_df['industry'].value_counts().sort_values(ascending=False)
# 转换为df，打印df_recommendation_stock_code
df_recommendation_industry = df_recommendation_industry.reset_index()
df_recommendation_industry.columns = ['industry', 'count']
df_recommendation_industry
```

```python
# 读取D:\workspace\xiaoyao\data\widetable.parquet文件，筛选date在2025-10-08以后的数据
import pandas as pd

file_path = r'D:\workspace\xiaoyao\data\widetable.parquet'
df = pd.read_parquet(file_path)
df = df[df['date'] == '2025-01-03']
```

```python

# 筛选出stock_code去除后缀后，仅获取前6位，在df_recommendation_stock_code['stock_code']中的数据
df = df[df['stock_code'].str[:6].isin(df_recommendation_stock_code['stock_code'])]
# 统计df的return_0d,return_1d的均值、众数，并打印屏幕
print(f"return_0d的均值为 {df['return_0d'].mean()}")
print(f"return_0d的众数为 {df['return_0d'].mode().values[0]}")
print(f"return_1d的均值为 {df['return_1d'].mean()}")
print(f"return_1d的众数为 {df['return_1d'].mode().values[0]}")
```
