```python
# 将../data目录下所有的文件拷贝到U盘。并显示进度条
import os
import shutil
from tqdm import tqdm

# 定义U盘挂载点
usb_drive = r'E:\xiaoyao_data\data'  # 根据实际挂载点修改

# 定义数据目录
data_dir = '../../data'

# 遍历数据目录下所有文件
for file_name in os.listdir(data_dir):
    file_path = os.path.join(data_dir, file_name)
    
    # 检查是否为文件（排除目录）
    if os.path.isfile(file_path):
        # 构建目标路径（U盘根目录）
        dest_path = os.path.join(usb_drive, file_name)
        
        # 拷贝文件到U盘
        shutil.copy2(file_path, dest_path)
        print(f"已拷贝 {file_name} 到 {usb_drive}")
```

```python
# 将../../data目录下的子目录stock_minutely_price先压缩为压缩文件。
# 拷贝到U盘指定目录E:\xiaoyao_data\data\stock_minutely_price后再解压缩。

import os
import shutil
import zipfile
from pathlib import Path

def compress_directory(source_dir, zip_filename):
    """
    压缩指定目录到ZIP文件
    """
    try:
        # 确保源目录存在
        if not os.path.isdir(source_dir):
            raise FileNotFoundError(f"源目录不存在: {source_dir}")
        
        # 创建ZIP文件
        with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
            # 遍历目录中的所有文件和子目录
            for root, dirs, files in os.walk(source_dir):
                for file in files:
                    file_path = os.path.join(root, file)
                    # 计算相对路径，保持目录结构
                    arcname = os.path.relpath(file_path, os.path.dirname(source_dir))
                    zipf.write(file_path, arcname)
        
        print(f"成功压缩目录到: {zip_filename}")
        return True
    except Exception as e:
        print(f"压缩失败: {str(e)}")
        return False

def copy_to_usb(zip_file, usb_dest_dir):
    """
    将ZIP文件复制到U盘指定目录
    """
    try:
        # 确保ZIP文件存在
        if not os.path.isfile(zip_file):
            raise FileNotFoundError(f"ZIP文件不存在: {zip_file}")
        
        # 确保目标目录存在，不存在则创建
        os.makedirs(usb_dest_dir, exist_ok=True)
        
        # 复制文件
        dest_path = os.path.join(usb_dest_dir, os.path.basename(zip_file))
        shutil.copy2(zip_file, dest_path)
        
        print(f"成功复制到U盘: {dest_path}")
        return dest_path
    except Exception as e:
        print(f"复制失败: {str(e)}")
        return None

def extract_zip(zip_file, extract_dir):
    """
    解压缩ZIP文件到指定目录
    """
    try:
        # 确保ZIP文件存在
        if not os.path.isfile(zip_file):
            raise FileNotFoundError(f"ZIP文件不存在: {zip_file}")
        
        # 确保提取目录存在
        os.makedirs(extract_dir, exist_ok=True)
        
        # 解压缩
        with zipfile.ZipFile(zip_file, 'r') as zipf:
            zipf.extractall(extract_dir)
        
        print(f"成功解压缩到: {extract_dir}")
        return True
    except Exception as e:
        print(f"解压缩失败: {str(e)}")
        return False

def main():
    # 配置路径
    source_dir = os.path.abspath("../../data/stock_minutely_price")
    zip_filename = os.path.abspath("./stock_minutely_price.zip")
    usb_dest_dir = "E:/xiaoyao_data/data/stock_minutely_price"
    
    # 步骤1: 压缩目录
    if not compress_directory(source_dir, zip_filename):
        return
    
    # 步骤2: 复制到U盘
    usb_zip_path = copy_to_usb(zip_filename, usb_dest_dir)
    if not usb_zip_path:
        return
    
    # 步骤3: 解压缩
    extract_zip(usb_zip_path, usb_dest_dir)
    
    # 可选: 删除临时ZIP文件
    if os.path.exists(zip_filename):
        os.remove(zip_filename)
        print(f"已删除临时ZIP文件: {zip_filename}")

if __name__ == "__main__":
    main()
```
