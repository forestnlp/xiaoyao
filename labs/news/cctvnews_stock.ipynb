```python
# 读取D:\workspace\xiaoyao\llm\news_final_verdict_20251010.csv。
import pandas as pd

df_news = pd.read_csv(r'D:\workspace\xiaoyao\llm\news_final_verdict_20251010.csv', dtype={'stock_code': str})
df_news.head()
```

```python
# 读取D:\workspace\xiaoyao\data\stock_daily_price.parquet
import pandas as pd

df_price = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_price.parquet')
df_price.head()
```

```python
# 定义日期为 2025-10-09 。读取 df_price 中该日期的数据。
date = '2025-10-09'
df_price_date = df_price[df_price['date'] == date]

# 计算return = (close-pre_close)/pre_close
df_price_date['return'] = (df_price_date['close'] - df_price_date['pre_close']) / df_price_date['pre_close']
df_price_date['return2'] = (df_price_date['close'] - df_price_date['open']) / df_price_date['open']

# 去掉stock_code里的后缀 仅保留六位股票代码
df_price_date['stock_code'] = df_price_date['stock_code'].str[:6]
df_price_date
```

```python
# 用df_news里的stock_code，查询df_price_date里对应的open\close\high\low\pre_close\return字段，df_news是主表，df_price_date是从表
df_news = df_news.merge(df_price_date[['stock_code', 'date', 'open', 'close', 'high', 'low', 'pre_close', 'return', 'return2']], on=['stock_code'], how='left')
df_news
```

```python
# 统计return>0比例，return的平均值，中位数，打印到屏幕
df_news['return'].mean()
df_news['return'].median()
```

```python
# 统计return>0比例，return的平均值，中位数，打印到屏幕
df_news['return2'].mean()
df_news['return2'].median()
```
