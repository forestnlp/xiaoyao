```python
import pandas as pd

# 分别读取量价信息、竞价信息、行业信息、新闻信息

# 仅读取2025年的量价信息
df_price = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_price.parquet')
df_price = df_price[df_price['date']>='2025-01-01']

df_industry = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_industry.parquet')

# 读取2025年的竞价信息,将date字段截取年月日，忽略时分秒。保持数据类型不变。
df_auction = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_auction.parquet')
df_auction['date'] = pd.to_datetime(df_auction['date']).dt.strftime('%Y-%m-%d')

df_market = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_marketcap.parquet')
df_news = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_daily_cctvnews.parquet')
df_minute = pd.read_parquet(r'D:\workspace\xiaoyao\data\stock_minutely_price.parquet')
```

```python

# 合并量价信息、竞价信息、行业信息
df = pd.merge(df_price, df_industry, on=['date', 'stock_code'], how='left')
df = pd.merge(df, df_market, on=['date', 'stock_code'], how='left')
df = pd.merge(df, df_auction, on=['date', 'stock_code'], how='left')
```

```python
# 将volume_x money_x 重命名为volume money
df = df.rename(columns={'volume_x': 'volume', 'money_x': 'money'})
# volume_y money_y 重命名为auc_volume auc_money
df = df.rename(columns={'volume_y': 'auc_volume', 'money_y': 'auc_money'})
```
